# __Univariate (one feature):__

```python
import pandas as pd

fraud_df = pd.read_csv('../data/raw/Fraud_Data.csv')
creditcard_df = pd.read_csv('../data/raw/creditcard.csv')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Age distribution
sns.histplot(fraud_df['age'])
plt.savefig('../reports/figures/age_distribution.png')
```

```python
# Gender distribution
sns.countplot(x='sex', data=fraud_df)
plt.title("Gender Distribution")
plt.savefig('../reports/figures/gender_distribution.png')
plt.show()
```

```python
# Gender distribution
sns.countplot(x='browser', data=fraud_df)
plt.title("Browser source")
plt.savefig('../reports/figures/browser.png')
plt.show()
```

# __Bivariate (two features):__

```python
# Fraud vs Non-Fraud by age
sns.boxplot(x='class', y='age', data=fraud_df)
plt.savefig('../reports/figures/age_vs_class.png')
```

```python
# Gender vs Fraud
sns.countplot(x='sex', hue='class', data=fraud_df)
plt.title("Gender vs Fraud")
plt.savefig('../reports/figures/gender_vs_fraud.png')
plt.show()
```

```python
# Gender vs Fraud
sns.countplot(x='browser', hue='class', data=fraud_df)
plt.title("Browser vs Fraud")
plt.savefig('../reports/figures/browser_vs_fraud.png')
plt.show()
```
