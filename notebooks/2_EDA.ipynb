# __Univariate (one feature):__

```python
import pandas as pd

fraud_df = pd.read_csv('../data/raw/Fraud_Data.csv')
creditcard_df = pd.read_csv('../data/raw/creditcard.csv')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Age distribution
sns.histplot(fraud_df['age'])
plt.savefig('../reports/figures/age_distribution.png')
```

# __Bivariate (two features):__

```python
# Fraud vs Non-Fraud by age
sns.boxplot(x='class', y='age', data=fraud_df)
plt.savefig('../reports/figures/age_vs_class.png')
```
